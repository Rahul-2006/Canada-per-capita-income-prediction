```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/kaggle/input/d/mfaisalqureshi/canada-per-capita-income/canada_per_capita_income.csv')
df.head(3)
```

```python
plt.scatter(x='year' , y='per capita income (US$)',data=df,marker='+',color='red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df[['per capita income (US$)']])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2020]])
```

```python
df.head()
```

```python
plt.scatter(x='year',y='per capita income (US$)',data=df,marker='+',color='red')
plt.plot(df.year,reg.predict(df[['year']]),color='green')
plt.xlabel('YEAR')
plt.ylabel('PER CAPITA INCOME(US$)')
plt.show()
```
